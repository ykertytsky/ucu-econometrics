# Short description of each econometrics data structure 
- Time-series
    One or multiple variables recorded over time for a single entity. For example, the monthly unemployment rate in a country from 2000 to 2020.
- Panel datag
    Data that combines both cross-sectional and time-series dimensions, involving multiple entities observed over multiple time periods. For example, annual income data for different households from 2010 to 2020.
- Cross-sectional data
    Data collected at a single point in time across multiple entities. For example, a survey of household incomes in a country conducted in 2020.
- Pooled Cross Sectional data
    Pooled cross-sectional data is cross-sectional dataset collected at different points in time. For example, survey data collected from different individuals in 2010 and then again in 2020.

# 2. Explain why you classify the dataset as a time series or a panel dataset, respectively. Clearly explain the difference
between the two.
This dataset can be classified as a panel dataset because it contains multiple observations (GDP values) for different entities (countries) over a series of time periods (years 2021 to 2025). A time series dataset typically involves a single entity observed over time, whereas a panel dataset involves multiple entities observed over the same time periods. In this case, since we have GDP data for multiple countries across several years, it fits the definition of a panel dataset.
